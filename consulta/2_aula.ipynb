```python
from shapely.geometry import  Point, LineString, Polygon
```

```python
ponto = Point(5, 10)
```

```python
list(ponto.coords)
```

```python
x = ponto.x
y = ponto.y
print(x, y)
```

```python
ponto1 = Point(2, 3)
ponto2 = Point(4, 5)
distancia = ponto1.distance(ponto2)
print(distancia)
```

```python
ponto1 = Point(2, 3)
ponto2 = Point(4, 5)
ponto3 = Point(7,9)
linha = LineString([ponto1, ponto2, ponto3])
linha
```

```python
list(linha.coords)
```

```python
linha.xy
```

```python
coordx = list(linha.xy[0])
coordy = list(linha.xy[1])
print(f'coordenadas x: {coordx}\ncoordenadas y: {coordy}')
```

```python
linha.length
```

```python
linha.centroid
```

```python
poligono = Polygon([(2, 4), (6, 6), (7, 10)])
poligono
```

```python
list(poligono.exterior.coords)
```

```python
area = poligono.area
print(f'A área do polígono é de {area:.0f} metros quadrados.')
```

```python
circulo1 = Point(1,1).buffer(1)
circulo2 = Point(2,1).buffer(1)

circulo2
circulo1
```

```python
x, y = circulo1.exterior.xy
plt.plot(x,y)
x, y = circulo2.exterior.xy
plt.plot(x,y)
plt.show
```

```python
circulo1.intersection(circulo2)
```

```python
circulo1.union(circulo2)
```

```python
circulo1.difference(circulo2)
```

```python
circulo1.symmetric_difference(circulo2)
```
